```python
df_train_copy = df_train.copy()

# Asignar valores numéricos a las categorías
category_mapping = {'bajo': 1, 'medio-bajo': 2, 'medio-alto': 3, 'alto': 4}
df_train_copy['RENDIMIENTO_GLOBAL_NUM'] = df_train['RENDIMIENTO_GLOBAL'].map(category_mapping)

# Graficar usando seaborn
sns.histplot(df_train_copy['RENDIMIENTO_GLOBAL_NUM'], bins=4, kde=False)

# Personalizar el gráfico
plt.xticks([1, 2, 3, 4], ['bajo', 'medio-bajo', 'medio-alto', 'alto'])
plt.xlabel('Rendimiento Global')
plt.ylabel('Frecuencia')
plt.title('Distribución de RENDIMIENTO_GLOBAL')
plt.show()
```

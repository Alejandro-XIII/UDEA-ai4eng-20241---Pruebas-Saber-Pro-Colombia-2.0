```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_validate, ShuffleSplit
from google.colab import drive
drive.mount("/content/drive/")
```

```python
#carga del archivo train.csv con pandas
df_train_pro = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BD/train_procesado.csv')
df_train_pro
```

```python
df_train_pro = df_train_pro.sample(len(df_train_pro))

d = df_train_pro.iloc[:300000].sample(1000)
X = d.values[:,1:-1]
y = d["RENDIMIENTO_GLOBAL"].values
print (X.shape, y.shape)
```

```python
print(X,y)
```

```python
def rel_mrae(estimator, X, y):
    preds = estimator.predict(X)
    return np.mean(np.abs(preds-y)/y)
```

```python
def report_cv_score(z):
    print ("test score   %.3f (±%.4f) with %d splits"%(np.mean(z["test_score"]), np.std(z["test_score"]), len(z["test_score"])))
    print ("train score  %.3f (±%.4f) with %d splits"%(np.mean(z["train_score"]), np.std(z["train_score"]), len(z["train_score"])))
```

```python
estimator1 = DecisionTreeRegressor(max_depth=4)
z1 = cross_validate(estimator1, X, y, return_train_score=True, return_estimator=True,
                    scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=0.1))
report_cv_score(z1)
```

```python
from sklearn.ensemble import RandomForestRegressor
estimator2 = DecisionTreeRegressor(max_depth=4)
z2 = cross_validate(estimator2, X, y, return_train_score=True, return_estimator=True,
                    scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=0.1))
report_cv_score(z2)
```

```python
from sklearn.linear_model import LinearRegression
estimator3 = LinearRegression()
z3 = cross_validate(estimator3, X, y, return_train_score=True, return_estimator=True,
                    scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=0.1))
report_cv_score(z3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
estimator4 = GradientBoostingRegressor()
z4 = cross_validate(estimator4, X, y, return_train_score=True, return_estimator=True,
                    scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=0.1))
report_cv_score(z4)
```

```python
from xgboost import XGBRegressor
estimator5 = XGBRegressor()
z5 = cross_validate(estimator5, X, y, return_train_score=True, return_estimator=True,
                    scoring=rel_mrae, cv=ShuffleSplit(n_splits=10, test_size=0.1))
report_cv_score(z5)
```

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount("/content/drive/")
```

```python
#carga del archivos
dtr = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BD/train_procesado.csv')
dts = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BD/test_procesado.csv')
dtr.shape, dts.shape
```

```python
dts.head()
```

```python
Xtr, ytr = dtr.values[:,1:-1], dtr["RENDIMIENTO_GLOBAL"].values
Xts      = dts.values[:,1:]

print (Xtr.shape, ytr.shape)
print (Xts.shape)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
rf = XGBRegressor()
```

```python
rf.fit(Xtr, ytr)
preds_ts = rf.predict(Xts)
preds_ts
```

```python
#Redondear al entero más cercano
preds_ts = np.round(preds_ts).astype(int)
preds_ts
```

```python
#Mapear los enteros a cadenas de texto
mapping = {1: "bajo", 2: "medio-bajo", 3: "medio-alto", 4: "alto"}
preds_ts = np.vectorize(mapping.get)(preds_ts)
preds_ts
```

```python
submission = pd.DataFrame([dts.ID, pd.Series(preds_ts, name="RENDIMIENTO_GLOBAL")]).T
submission.head()
```

```python
submission.to_csv('/content/drive/MyDrive/Colab Notebooks/BD/submission.csv', index=False)
```
